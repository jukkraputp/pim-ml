# Preprocessing

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
X = [['B'], ['A'], ['C'], ['D']]
```

```python
encoder.fit(X)
```

```python
X_ = encoder.transform(X)
```

```python
X_
```

```python
encoder.categories_
```

```python
X = [['A', 10], ['B', 9.5], ['C', 8], ['D', 16]]
```

```python
encoder.fit(X)
```

```python
X_ = encoder.transform(X)
```

```python
X_
```

```python
encoder.categories_
```

```python
encoder.inverse_transform(X_)
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
Y = ['bat', 'ant', 'cat', 'dog', 'ant']
```

```python
encoder.fit(Y)
```

```python
Y_ = encoder.transform(Y)
```

```python
Y_
```

```python
encoder.classes_
```

```python
encoder.inverse_transform(Y_)
```

# One-hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
X = [['ant'], ['bat'], ['cat'], ['bat'], ['ant'], ['cat']]
```

```python
encoder.fit(X)
```

```python
X_ = encoder.transform(X)
```

```python
X_
```

```python
X_.todense()
```

```python
encoder.categories_
```

```python
encoder.inverse_transform(X_)
```

```python
df
```

```python
pd.get_dummies(df)
```
